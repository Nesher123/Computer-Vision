Read image into an array im1 and dispaly the array as an image

```python
im1 = cv2.imread('yami.jpg')
cv2.imshow('Color Image',im1)
```

As mentioned on Piazza, while using Jupyter NoteBook (and might be with some other ineterpreter), you should use **cv.waitKey(0)** to show an image. That happens since the image box is using the same Python process as the kernel. When execute this from interactive shell, the shell waits for you to get back from the image display.
This sometimes bring Jupyter to crash, so there are the option to add the function: **cv2.destroyAllWindows()** at the end of the run, or to use plt instead as shown below.

```python
cv2.waitKey(0) #wait for any key
cv2.destroyAllWindows()
```

Using plt from matplotlib for inline results (Jupyter notebook):

```python
im1_gray = cv2.imread('yami.jpg', cv2.IMREAD_GRAYSCALE)

imgPlot = plt.imshow(im1_gray, cmap = 'gray')
plt.title('yami.jpg in gray levels')
```

Convert to gray scale image

```python
im1_gray=cv2.cvtColor(im1, cv2.COLOR_BGR2GRAY)
im1_gray.shape
cv2.imshow('Gray Image',im1_gray)

cv2.waitKey(0)
cv2.destroyAllWindows()

cv2.imwrite('yami_G.jpg',im1_gray)
```

Use a threahold to obtain a binary image

```python
im3=cv2.imread('yami_G.jpg',cv2.IMREAD_GRAYSCALE)
print(im3.shape)


th=180 
im1_gray[im1_gray>th]=255
im1_gray[im1_gray<=th]=0
cv2.imshow('Binary image',im1_gray)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
